```python
import pymongo
```

```python
dbConn=pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
dbname="demo_db"
db=dbConn[dbname]
```

```python
#show all databses
print(dbConn.list_database_names())
#note if db has no data it wont be listed, to check store this in a list and check for db
```

```python
#connecting to collection i.e tables
print(db.list_collection_names())
```

```python
#connecting to a cl=ollection
collection_name='mongo'
collection=db[collection_name]
```

```python
#retrieving all records
result=collection.find({})
result[3]
#note find gives all docs which may lead to loading issues , so always use limit() to limit the rows 
```

```python
result_total=collection.find({}).limit(3)
for res in result_total:
              print(res)
```

```python
result_some=collection.find({},{'GRE Score','TOEFL Score'})

for res in result_some:
    print(res)
```

```python
#find rows in specific criteria

my_db_query={'GRE Score':'337'}

result_GRE=collection.find(my_db_query)
for res in result_GRE:
    print(res)
```

```python
#greater than is metioned as $gt
my_db_query ={'GRE Score':{'$gt':'337'}}

result_GRE=collection.find(my_db_query)

for res in result_GRE:
    print(res)
```

```python
result_GRE= collection.find(my_db_query).sort('GRE Score')
#default is asc order
for res in result_GRE:
    print(res)
```

```python
#to sort in desc order
result_GRE= collection.find(my_db_query).sort('GRE Score',-1)

for res in result_GRE:
    print(res)
```

```python
#delete rows
my_db_query={'GRE Score':'337'}

x=collection.delete_one(my_db_query)
print(x.deleted_count)

#for deleteing many collection.delete_many(my_db_query) 
#for deleting all docs in a collection collection.delete_many({})
```
